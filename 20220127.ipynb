```python
score = int(input("당신의 점수를 입력하세요 >>> "))

#int는 인덱싱 불가

if score >= 90:
  grade = ('A')
elif score >= 80:
  grade = ('B')
elif score >= 70:
  grade = ('C')
elif score >= 60:
  grade = ('D')
else :
  grade = ('F')

if int(score%10) >= 7:
  grade2 = ('+')
elif int(score%10) >= 4:
  grade2 = (' ')
else :
  grade2 = ('-')

print(f'당신의 점수는 {score}이고 학점은 {grade}{grade2} 입니다.')

```

```python
###############################################3
# While
##################################################

#반복문은 작업 수행을 한 번이 아니라 계속해서 수행해야 할때 사용
#while 반복 작업을 몇 번 수행할지 확실하지 않을때 주로 사용
#ex) 운동장을 해가 질때까지 반복해서 돌자

# while 조건식 :
# 실행구문

#while True:
#  print('운동장 돌기')
```

```python
n = 1
while n <= 10 :
  print(n)
  n+=1 #n = n + 1 같음 (자바 n++)

print('---------------')
print(n)
```

```python
#10부터 1까지 짝수만 거꾸로 출력

n = 10
while n>=1 :
  print(n)
  n-=2
```

```python
# 사용자보루터 임의의 정수를 입력받아 모두 리스트에 보관합니다.
#단 0을 입력하면 프로그램을 종요합니다. (0은 리스트에 저장 안함)
# 촤종 저장된 리스트를 출력

nlist = []
number = 1

while number != 0:
  number = int(input('>> 임의의 정수를 입력하세요 '))

  if number != 0:
    nlist.append(number)
  
  


print(nlist)

```

```python
# while 중첩
#while 내부에 또 다른 while을 사용하는 것을 중첩이라고 한다.

# 3일 동안 1~3교시 수업을 한다.

day = 1
class1 = 0

while day <= 3 :
  class1 += 1
  while class1 <= 4 :
    day += 1
    print(f'{day}일차 {class1}교시 입니다.')


```
